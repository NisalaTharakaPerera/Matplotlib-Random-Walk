## Now let's create a Random Walk

```python
step_option = [1,-1]
step_choice = choice(step_option)
step_choice   # This will randomly select from 1 or -1 everytime it is called
```

```python
walk = []
walk.append(step_choice)
for step in range(1,1000):
    next_step = walk[step-1] + choice(step_option)
    walk.append(next_step)
print(walk)
```

```python
plt.plot(walk)
plt.show()
```

## Doing the same using a function

```python
def rand_walk(step_num):
    walk = []
    step_choice = choice([1,-1])
    walk.append(step_choice)
    for i in range(1,step_num):
        step_choice2 = choice([-1,1])
        next_step = walk[i-1] + step_choice2
        walk.append(next_step)
    return walk
```

```python
random_w = rand_walk(100)
print(random_w)
```

```python
def plot_walk(walk):
    plt.plot(walk)
    plt.xlabel("Number of steps")
    plt.ylabel("Distance from origin")
    plt.title("OUr Random Walk")
    plt.show()
```

```python
plot_walk(rand_walk(1000))
```

## We can create a function to plot more than one random walk

```python
def plot_walks(num,walk_steps):
    
    lab_list = list(range(1,num+1))  #label list
    
    for i in range(0,num):
        
        x = list(range(1,walk_steps+1))
        
        plt.plot(x,rand_walk(walk_steps),label = "Plot Number" + ' ' + str(lab_list[i]))
        plt.xlabel("Number of Steps")
        plt.ylabel("Distance from Origin")
        plt.title("Our Random Walk")
        plt.legend(loc="lower left")  # a legend tells which plot is which
    
    plt.show()       
```

```python
plot_walks(5,100000)
```
